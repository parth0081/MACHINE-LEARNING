```python
import seaborn as sns
import numpy as np
```

```python
df=sns.load_dataset('tips')
df.head()
```

# CATEGORICAL PLOTS



#### Countplot (uses count on y-axis and 1 or 2 categorical features)

```python
sns.countplot(x='smoker',data=df)
```

```python
sns.countplot(x='smoker',data=df, hue='sex')
```

#### BAR PLOT (1 categorical and 1 numerical feature)

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
sns.barplot(x='total_bill',y='sex',data=df)
```

```python
sns.barplot(x='sex',y='total_bill',data=df,estimator=np.median)   #median value on y-axis
```

```python
sns.barplot(x='sex',y='total_bill',data=df,estimator=np.mean)
```

#### BOXPLOT (uses 1 or 2 categorical and 1 numerical)

```python
sns.boxplot(x='sex',y='total_bill',data=df)
```

```python
sns.boxplot(x='sex',y='total_bill',hue='smoker',data=df)
```

#### HUE CANNOT TAKE MORE THAN ONE FEATURE

```python
sns.boxplot(x='sex',y='total_bill',hue=['smoker','time'],data=df)
```

#### VIOLINPLOT (uses 1 or 2 categorical and 1 numerical)

```python
sns.violinplot(x='sex',y='total_bill',data=df)
```

```python
sns.violinplot(x='sex',y='total_bill',hue='smoker',data=df)
```

```python
sns.stripplot(x='sex',y='total_bill',data=df)
```

```python
sns.stripplot(x='sex',y='total_bill',hue='smoker',data=df)
```

```python
sns.swarmplot(x='sex',y='total_bill',data=df)
```

```python
sns.swarmplot(x='sex',y='total_bill',hue='smoker',data=df)
```
